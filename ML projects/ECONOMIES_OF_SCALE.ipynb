```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
economic=pd.read_csv('EconomiesOfScale.csv')
```

```python
economic.info()
```

```python
sns.jointplot(x='Number of Units',y='Manufacturing Cost',data=economic)
```

```python
sns.jointplot(x='Number of Units',y='Manufacturing Cost',data=economic,kind='hex')
```

```python
sns.lmplot(x='Number of Units',y='Manufacturing Cost',data=economic)
```

```python
X=economic[['Number of Units']]
y=economic['Manufacturing Cost']
```

```python
X_train=X
y_train=y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regressor=LinearRegression(fit_intercept=True)
```

```python
regressor.fit(X_train,y_train)
```

```python
print('Linear Model Coefficent (m):',regressor.coef_)
print('Linear Model Coefficent (b):',regressor.intercept_)
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regressor=PolynomialFeatures(degree=4)
```

```python
X_columns=poly_regressor.fit_transform(X_train)
```

```python
print(X_columns)
```

```python
regressor_poly=LinearRegression()
regressor_poly.fit(X_columns,y_train)
```

```python
print('Model Coefficent: ',regressor_poly.coef_)
print('Model Coefficent: ',regressor_poly.intercept_)
```
